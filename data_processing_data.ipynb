```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy='mean')
imputed = imputer.fit_transform(df[['Age','Salary']])
imputed = np.round(imputed,2)
df[['Age','Salary']] = imputed
```

```python
df.iloc[:,0]
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
col=['Purchased']
df[col] = imputer.fit_transform(df[col])
df
```

```python
from sklearn.preprocessing import LabelEncoder 
labelencode = LabelEncoder()
df['Purchased'] = labelencode.fit_transform(df['Purchased'])
df
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
df[['Age','Salary']] = sc.fit_transform(df[['Age','Salary']])
df
```

```python
from sklearn.preprocessing import LabelEncoder
labelencode = LabelEncoder()
df['Country'] = labelencode.fit_transform(df['Country'])
df
```
